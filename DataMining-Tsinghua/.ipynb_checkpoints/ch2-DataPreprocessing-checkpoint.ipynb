## Data Transformation

**Attribute Types**
- Continuous
- Discrete：离散的
- Ordinal：打分
- Norminal：枚举类型，如何编码  RGB 0 1 2 编码正确吗？默认空间距离不同
- String：文本数据，如何对数据进行编码

**Type Conversion**
- 数据编码：会导致空间信息改变，如RGB 0 1 2
- 独热编码(onehot)


**Sampling**
- 数据采样降低时间和空间复杂度
- 大数据：挑一些出来   统计学：采样
- Aggregation 数据聚合，降低数据量
- 重要作用：调整类的比例，防止训练集不平衡，不能只看准确率

**Imbalanced Datasets**
- 准确率
- 召回率


**Over-Sampling**

数据集严重不平衡：把数据少的多复制生成，思想是在相邻的区域内随机生成一些数据

**Boundary-Sampling**

边界采样。有时候**边缘点**的意义比较大，而内部数据并大重要。节省大量计算资源

**Normalization** 标准化
> 计算机中数据没有单位，标准化没有影响，映射到0-1区间
- Min-max normalization：有明确的上下限
- Z-score normalition：该点偏离均值多远

## Feature Selection  特征选择

**Class Distributions** 属性的好坏 分布的概率

**Entropy** 熵  先验概率  --- 量化区分度
$$H(X) = \sum_{i=1}^n p(x_i)log_bp(x_i)$$



**Information Gain**:信息增益，当你知道一个额外的属性时，系统的不确定性降低了多少。该值越大，说明属性越重要

**Feature Subset Search**： 找到最优的属性组合 --> 组合爆炸问题
- Branch and Bound 分支限界算法，降低搜索解空间


局部最优解：

- Top K Individdual Features
- Sequential Forward Selection: 先找到一个最优属性，然后依次添加一个属性，找到当前最优组合
- Sequential Backward Selection：先使用所有属性，然后依次去除某一属性


Optimization Algorithms
- Simulated Annealing
- Tabu Search
- Genetic Algorithms
