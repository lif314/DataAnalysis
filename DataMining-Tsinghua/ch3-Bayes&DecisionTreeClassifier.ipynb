## Decision Tree

- 同一数据集可以生成不同的决策树
- 可解释性强，符合人类思维和规则
- Occam's Razor(剃刀): 可以把不需要的东西剔除，剪枝

### ID3(Iterative Dichotomizer 3)
- 如何建造小的、短的决策树
    - 选择属性好的属性
    - 强的属性在上层
    - Ross Quinlan
    
- Entropy(熵)
    - 衡量系统(变量)的不确定性
    - $S$样本
    - 计算原始数据集的熵(不确定性)，加入一个属性，划分后计算新的熵，计算两者之间的差(信息熵 Information Gain),以此来选择好的属性
    - Information Gain越大越好


- ID3 Framework
    - 递归结构
    - 一个属性用过则不能再用，如果分到最后依然不纯，则少数服从多数

    - 实例：选择属性，纯不纯？继续选择继续划分
    

**Overfitting**
- 决策树太长会导致overfitting
- Definition: 给定A和B两个分类器，A在训练集的误差比B要小，但在测试集中误差比B高，则A是过拟合

- Solutions:
    - Stop growing the tree earilier 限制树的长度
    - Allow the tree to overfit the data and then post-prune(剪枝) the tree

**post-prune**
- 从上往下剪枝，不是剪去，而是合并，少数服从多数。
- 三个数据集：training data, testing data, validation data
    - 在traing data（树疯狂长）,开始剪枝时，需要参照模型在validation data上的performance：刚开始，在校验集上的误差会逐渐较低，剪得太多，模型太简单，误差会上升，在拐点处停止剪枝。
    - 模型太复杂，过拟合；模型太简单，不足以描述数据的特征

**Entropy Bias(熵偏差)**
- 属性：有些属性不适合做分类，把样本分的太细，容易过拟合
- penalized: 评判属性，切的越细，越大(S,A)， 惩罚量越大

- 离散化的Threshold, 计算信息熵进行评判
