## Non-linear SVMs

- 在一维空间线性不可分的，将其映射在二维空间(Feature Space)中


- 分析关系，坐标映射


- 映射在三维空间


如何做映射？实际上并不需要为每一问题单独做一个映射。可以使用固定的几种映射方式。

为什么要升维？降维是降低问题的复杂度，升维？

**Quadratic Basis Functions**


### SVM升维

- SVM最核心操作是两个向量做内积


- 高维空间的内积简化：


- **最著名的特性**
    - 高维空间中的操作等价于低维空间中的操作
    - $O(m^2)\rightarrow O(m)$
$$K(a, b) = (a\cdot b + 1)^2 = \Phi(a) \cdot \Phi(b)$$

### Kernel Trick

我滴乖儿！！！
- 利用了高维空间的分类效果
- 又巧妙地绕过了高维空间的计算量


### Kernels

- 核函数需要满足一定的特性
- 映射到多项式  $\frac{m^2}{2}$维
- 高斯映射到无穷维，但不用关心，可以在低维进行计算


### String Kernel

- 核函数：线性分类器处理线性不可分问题
- 处理文本数据：文本怎样变为数据
- 计算文本中的字串
- 包含cat字串，包含c-a映射为$\lambda^2$, c-t中是cat长度为3，映射为$\lambda^3$
- 文本在高维空间做内积


- 求$w,b$: 全部加$\Phi$,但不需要计算出来，都是在做内积，可以在低维上运算
