### Perceptrons（感知机、神经元）

- input: 特征 $x_1, x_2, ..., x_n$
- 权重： $w_1, w_2, ..., w_n$
- cell：内积 $\sum_{i=0}^nw_i\cdot x_i$
- $w_0$: $x_0$始终等于1，相当于一个额外的偏值(bias)。实质是形成判别平面，否则会一直经过原点.会导致不收敛
- 感知机就是一个超平面，共n+1个输入参数



- 门限：将输入转化为0或1


**Gradient Descent** 梯度下降

- 误差: $t_d-t_O$  多个样本累加计算整体误差--> Batch Learning（批处理）
- 对误差求偏导，梯度信息(调整方向)
- 学习率：调整的幅度
- 负号，误差($E$)对权重($W$)求偏导,如果是正的，则说明随着w的增加，误差在增大。为了降低误差，就要减小$w$的权重，所以取负号

- Delta Rule

- 假设节点是线性的，即输入等于输出

**Batch Learning**
- 批处理：每来一条样本，计算误差然后更新$\Delta w_i$, 最后才更新$W_i$


**Stochastic Learning** 随机学习
- 来一条样本，就计算误差，结算结束后就立即做出调整
- 线性不可分问题&线性可分问题


## Multilayer Perceptron 多元感知机

### XOR问题：线性不可分问题

- 使用两根线/椭圆划分：两个决策面

- 使用两个决策面
    - OR：或门
    - NAND：与非门
    - AND：与门


### Hidden Layer Representation

- Hidden Layer: 
    - 输入层与输出结果是线性不可分的，没法使用一条线进行区别
    - 解决：先将输入层映射在隐含层上，问题变为线性可分问题，然后使用感知机(与门)进行划分。


### The Sigmoid Threshold(激活函数) Unit

- 保证输出始终在$[0,1]$内，映射很好
- 可以求导，导数可以直接计算(输入越接近0，导数最大)


## Backpropagation Rule 逆向传播

- 基本思想：误差对权重求偏导，
- Downstream(j)：使用j作为输入的下游节点

### Training Rule for Output Units

- 输出层：$o_j(1-o_j)$激活函数是sigmoid，与感知机计算一样
- 使用sigmoid函数，导数可计算


### Traing Rule for Hidden Units

- 不知道隐藏层的输出，不知道怎么计算误差
- 解决思路：将误差逆传播回来。以及误差的权重(影响力，关系)
- Downstream(j)：使用j作为输入的所有下游节点。误差汇总在j上
- 整体思路：先把可以计算误差的输出层进行误差计算，然后按照downstream(j)将误差和误差权重进行逆向传播，然后在下轮前向训练中进行调整


### BP Framework

**局部最优点问题**
- 再试一次
- 进化神经网络
- 神经网络的表达能力：可以逼近任意的复杂函数
- 深度学习：多多个感知机


**过拟合问题**
- 训练集，在训练过程中使用校验集进行校验，在转折点停止学习

**陷入局部最优**
- 添加一个冲量
- 学习率大小：0.05等。太小会get stuck，太大会跳过最优解，反复横跳

### Beyond BP Networks

### Elman Network 处理XOR问题

- 输入：01串，每次只输入1位。只有一个输入单元
- 输出：? dont't care。上下文关系：输入0不知道干什么，输入1可以预测你要做什么
- 记忆功能：copy, t-1时刻将蓝色节点copy到绿颜色节点中


### Hopfield Network

- 全互联网络，不知道输入输出
- 联想记忆？pattern输入，存入某个节点，下一次输入，网络收敛进行查找，找到相似的pattern
