## Sequential Leader Clustering

- 适用于流数据，一个一个处理
- 不需要k值，不需要迭代
- 基本原理：新来的点和原来的点计算距离，按照原先设置的阈值判断新的点是加入还是丢弃。如果加入，则需要重新计算中心点，然后处理下一个点
- Threshold: 太大，一个簇；太小，簇太多

## K-means中的思想

**K-Means Revisited**



- K-means的理论思想：两个变量都不知道，但循环中知道一个可以求另外一个，先假设参数，迭代计算调整[局部最优解]
- $\theta$模型参数：两个中心点的坐标，真正想要解决的问题
    - 一堆数据中无法找到中心点
- $Z$是每个点属于哪个簇，隐含参数[不需要计算，可以辅助解决问题]。求不出$Z$
- 知道$\theta$就知道$Z$，知道$Z$就可以计算$\theta$



## EM算法(期望最大化算法)

**最大似然估计**
- 知道抛的硬币是A和B
- H：正面
- T：反面


**EM算法**
- 不知道抛的是哪个硬币，怎么猜出$\theta_A$和$\theta_B$(A和B向上的概率)

- 工作流程：
    - 先猜测$\theta_A$和$\theta_B$在模型中计算，即知道了A和B向上的概率
    - 使用Bayes概率计算抛的是哪个硬币(A还是B)
    - 然后根据抛A和B的概率进行分配，则可以计算抛A和B向上的概率
    - 与假设的结果有差别，可以进行**学习**，直到收敛


- 只能保证局部最优

## EM: Gaussian Mixture

- 假设一维高斯
- 空间中任何一个点都可能由任意一个高斯函数生成
- 目标参数：均值和权重
- 辅助变量$Z_{ij}$：第i个样本是从第j个高斯中生成的，概率值。一个“跳板”
- $\theta$: $\alpha$ $\mu$

- **计算流程**：刚开始设置一些$\alpha$和$\mu$,则可以计算$z_{ij}$,然后加权后可以计算处出$\alpha$和$\mu$，迭代计算获得局部最优解
