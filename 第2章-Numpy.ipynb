**将任何一个nan替换为均值**

```python
a = np.arange(1,13).reshape(3,4).astype("float")
a[1, 3] = np.nan
print(a)
```

```python
# 遍历每一列
def fill_nan_mean_ndarray(a):
    for i in range(a.shape[1]):
        temp_col = a[:, i]  # 处理当前一列
        nan_num = np.count_nonzero(temp_col!=temp_col) # 不等于说明为nan -- 布尔索引
        if nan_num != 0:  # 当前列中存在nan
            temp_not_nan_col = temp_col[temp_col==temp_col]  # 当前列不为nan的array
            # 选中当前nan的位置，把均值填充进去
            temp_col[np.isnan(temp_col)] = temp_not_nan_col.mean()
    return a
```

```python
a = fill_nan_mean_ndarray(a)
print(a)
```
