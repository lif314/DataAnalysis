## 1.1 折线图

```python
from matplotlib import pyplot as plt # 导入pyplot


# matplotlib默认不支持中文，使用中文
# import  matplotlib
#  matplotlib.rc = ""

# axis  x轴
```

- 绘制简单图片

## 1.2 绘制散点图

```python
x  = range(2,26,2)
y = [12,23,34,45,48,51,53,56,61,71,82,88]
y_1 = [100,87,78,74,61,56,50,32,22,25,15,10]
```

```python
# 设置图形大小
plt.figure(figsize=(20, 8), dpi=80)

# 绘制散点图
plt.scatter(x, y, label="y", color="red")
plt.scatter(x, y_1, label="y_1")

# 必须要使用才能显示labeel
plt.legend()

# 设置刻度
plt.xticks(range(2, 25, 2))
plt.yticks(range( min(y), max(y) + 1, 5))


#  设置标题
plt.title("Scatter(x, y)")
plt.xlabel("time")
plt.ylabel("temperature")

plt.show()
```

## 1.3 绘制条形图

```python
a = ['a', 'b', 'c', 'd', 'e']

b = [123, 34, 45, 67, 23]

c = [12, 45, 56, 78, 99]
```

```python
# 设置图像大小
plt.figure(figsize=(20,8), dpi=80)

# 条形图
plt.bar(range(len(a)), b, width=0.3, label="blue")

#  需要向右移动一定的[width]距离,在进行绘制，否则会被覆盖
c_a =  [i + 0.3 for i in list(range(len(a)))]
plt.bar(c_a , c, width=0.3, label="orange")

plt.legend()


# 设置x轴显示
_xticks = a
plt.xticks(range(len(a)), _xticks)

# 横向的条形图   barh

plt.show()
```

```python
# 设置图像大小
plt.figure(figsize=(20,8), dpi=80)

# 条形图
plt.barh(range(len(a)), b, height=0.3)
plt.barh(c_a, b, height=0.3)

# 设置x轴显示
_yticks = a
plt.yticks(range(len(a)), _yticks)

# 横向的条形图   barh

plt.show()
```

## 1.4 直方图

```python
# 直方图：统计在某区间的数据

import random
x= [random.randint(0, i) for i in range(100)]

# 计算组数
d = 5 # 组距

num_bins = int((max(x) - min(x)) // d)

plt.hist(x, num_bins) #  x + 组数

plt.grid(5)

# 设置刻度
plt.xticks(range(min(x), max(x) + d, 5))

plt.title("hist graph")

plt.show()
```
